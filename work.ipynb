```python
from pdf2image import convert_from_path
import os

# Đường dẫn đến file PDF
pdf_path = "Vo o li Toan KN 3.1 ma QR.pdf"
output_folder = "images"

# Tạo thư mục để lưu ảnh nếu chưa có
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Chuyển PDF thành ảnh
images = convert_from_path(pdf_path)

# Lưu từng ảnh
for i, image in enumerate(images):
    image.save(f"{output_folder}/page_{i+1}.jpg", "JPEG")
```

```python
import os
import random
import shutil

# Tạo thư mục cho tập train, val và test
os.makedirs('project-7.1/images/train', exist_ok=True)
os.makedirs('project-7.1/images/val', exist_ok=True)
os.makedirs('project-7.1/images/test', exist_ok=True)
os.makedirs('project-7.1/labels/train', exist_ok=True)
os.makedirs('project-7.1/labels/val', exist_ok=True)
os.makedirs('project-7.1/labels/test', exist_ok=True)

# Liệt kê tất cả các file ảnh
image_files = [f for f in os.listdir('project-7.1/images') if f.endswith('.jpg')]
random.shuffle(image_files)

# Bước 1: Chia thành tập train+val và test
test_ratio = 0.25  # 20% dữ liệu cho test
test_count = int(len(image_files) * test_ratio)

train_val_files = image_files[:-test_count]  # 80% cho train+val
test_files = image_files[-test_count:]       # 20% cho test

# Bước 2: Chia tập train+val thành train và val
val_ratio = 0.2  # 20% của tập train+val cho validation
val_count = int(len(train_val_files) * val_ratio)

train_files = train_val_files[:-val_count]  # 80% của train+val cho train
val_files = train_val_files[-val_count:]    # 20% của train+val cho val

print(f"Tổng số ảnh: {len(image_files)}")
print(f"Số ảnh train: {len(train_files)} ({len(train_files)/len(image_files)*100:.1f}%)")
print(f"Số ảnh validation: {len(val_files)} ({len(val_files)/len(image_files)*100:.1f}%)")
print(f"Số ảnh test: {len(test_files)} ({len(test_files)/len(image_files)*100:.1f}%)")

# Di chuyển các file vào các thư mục tương ứng - Tập train
for f in train_files:
    # Di chuyển ảnh
    shutil.copy(f'project-7.1/images/{f}', 
                f'project-7.1/images/train/{f}')
    # Di chuyển label (nếu tồn tại)
    label_file = f.replace('.jpg', '.txt')
    if os.path.exists(f'project-7.1/labels/{label_file}'):
        shutil.copy(f'project-7.1/labels/{label_file}', 
                    f'project-7.1/labels/train/{label_file}')

# Di chuyển các file vào các thư mục tương ứng - Tập validation
for f in val_files:
    # Di chuyển ảnh
    shutil.copy(f'project-7.1/images/{f}', 
                f'project-7.1/images/val/{f}')
    # Di chuyển label (nếu tồn tại)
    label_file = f.replace('.jpg', '.txt')
    if os.path.exists(f'project-7.1/labels/{label_file}'):
        shutil.copy(f'project-7.1/labels/{label_file}', 
                    f'project-7.1/labels/val/{label_file}')

# Di chuyển các file vào các thư mục tương ứng - Tập test
for f in test_files:
    # Di chuyển ảnh
    shutil.copy(f'project-7.1/images/{f}', 
                f'project-7.1/images/test/{f}')
    # Di chuyển label (nếu tồn tại)
    label_file = f.replace('.jpg', '.txt')
    if os.path.exists(f'project-7.1/labels/{label_file}'):
        shutil.copy(f'project-7.1/labels/{label_file}', 
                    f'project-7.1/labels/test/{label_file}')
```

```python
yaml_content = """
# Đường dẫn tới thư mục gốc dataset
path: d:/Work/Ultralystic/project-7.1 
train: images/train  # Thư mục ảnh train
val: images/val      # Thư mục ảnh validation
test: images/test    # Thư mục ảnh test

# Số lượng lớp
nc: 6

# Tên các lớp
names: ['Nối', 'Trắc nghiệm', 'Tự luận', 'Điền từ', 'Đặt tính rồi tính','None']
"""
# path: sửa lại theo đường dẫn của bạn

with open('project7.yaml', 'w', encoding='utf-8') as f:
    f.write(yaml_content)

print("✅ Đã tạo file cấu hình YAML")
```

```python
from ultralytics import YOLO
import torch
```

```python
# Kiểm tra GPU
if torch.cuda.is_available():
    print(f"Sử dụng GPU: {torch.cuda.get_device_name(0)}")
    device = 0  # Sử dụng GPU đầu tiên
else:
    print("Không tìm thấy GPU, sử dụng CPU")
    device = 'cpu'
```

```python
# Tải mô hình YOLOv8 pretrained
try:
	model = YOLO('yolov8m.pt')  # Bạn có thể chọn các kích thước khác: s (small), m (medium), l (large), x (xlarge)
except RuntimeError as e:
	if "PytorchStreamReader failed reading zip archive" in str(e):
		print("❌ Model file 'yolov8m.pt' is corrupted or missing. Re-downloading...")
		import torch
		torch.hub.download_url_to_file('https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8m.pt', 'yolov8m.pt')
		model = YOLO('yolov8m.pt')  # Retry loading the model
	else:
		raise e
```

```python
results = model.train(
    data='project7.yaml',
    epochs=50,
    imgsz=960,
    batch=16,
    patience=20,
    device=device,
    workers=4,
    verbose=True,
    project='project7_results',
    name='yolov8m',
    exist_ok=True,
    pretrained=True,
    optimizer='AdamW',
    lr0=0.002,
    lrf=0.01,
    momentum=0.937,
    weight_decay=0.0005,
    warmup_epochs=3,
    warmup_momentum=0.8,
    warmup_bias_lr=0.1,
    box=5.0,
    cls=1.0,
    dfl=2.0,
    nbs=64,
    overlap_mask=True,
    save=True,
    save_period=-1,
    plots=True
)
```

```python
model = YOLO("project7_results/yolov8s/weights/best.pt")
model.val(data="project7.yaml", imgsz=640, batch=16, device="cuda" if torch.cuda.is_available() else "cpu", split="test")
```
